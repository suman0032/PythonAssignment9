```python
dt='https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv'
import pandas as pd
data = pd.read_csv(dt)
```

```python
data_new= data.loc[:,~data.columns.str.contains('^Unnamed')] #Question 1: removed unnamed columns
data_new
```

```python
gender_dist=pd.Categorical(data_new['Gender'])  #Question 2: Distribution of male and female
gender_dist=gender_dist.rename_categories(['Female','Male'])
gender_dist.describe()
```

```python
#Question 3 : the top 5 most preferred names
name_dist=pd.Categorical(data_new['Name']).describe()
name_dist.sort_values(by=['counts'],ascending=0)[:5]
```
